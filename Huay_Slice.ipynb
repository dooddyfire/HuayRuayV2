```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://www.ruay.at/%e0%b8%95%e0%b8%a3%e0%b8%a7%e0%b8%88%e0%b8%ab%e0%b8%a7%e0%b8%a2%e0%b8%ae%e0%b8%b2%e0%b8%99%e0%b8%ad%e0%b8%a2/"
res = requests.get(url)
soup = BeautifulSoup(res.content,'html.parser')

col1_lis = []
col2_lis = []
col3_lis = []
col4_lis = []

for row in soup.find_all('tr')[5:]:
  td_lis = [ i.text for i in row.find_all('td')]
  print(td_lis)
  col1_lis.append(td_lis[0])
  col2_lis.append(td_lis[1])
  col3_lis.append(td_lis[2])
  col4_lis.append(td_lis[3])

df = pd.DataFrame()
df['หวยฮานอยงวดวันที่'] = col1_lis
df['เลข 4 ตัว'] = col2_lis
df['เลข 3 ตัวบน (ร้อย)'] = [ str(i)[0] for i in col3_lis]
df['เลข 3 ตัวบน (สิบ)'] = [ str(i)[1] for i in col3_lis]
df['เลข 3 ตัวบน (หน่วย)'] = [ str(i)[2] for i in col3_lis]
df['เลข 2 ตัวล่าง (สิบ)'] = [ str(i)[0] for i in col4_lis]
df['เลข 2 ตัวล่าง (หน่วย)'] = [ str(i)[1] for i in col4_lis]

df.to_excel('Test_full.xlsx')
```
